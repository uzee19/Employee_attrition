```python
from sklearn.preprocessing import LabelEncoder
le_b=LabelEncoder()
le_d=LabelEncoder()
le_j=LabelEncoder()

le_gender=LabelEncoder()
le_marry=LabelEncoder()
le_overtime=LabelEncoder()
le_education_field=LabelEncoder()
```

```python
df['business_n']=le_b.fit_transform(df['BusinessTravel'])
df['department_n']=le_d.fit_transform(df['Department'])
df['job_n']=le_j.fit_transform(df['JobRole'])

df['gender_n']=le_gender.fit_transform(df['Gender'])
df['marry_n']=le_marry.fit_transform(df['MaritalStatus'])
df['overtime']=le_overtime.fit_transform(df['OverTime'])
df['EducationField_n']=le_education_field.fit_transform(df['EducationField'])
df.head()

```

```python
X=df[['EducationField_n','PerformanceRating','PercentSalaryHike','DistanceFromHome','Education','EmployeeNumber','Age','EnvironmentSatisfaction','JobInvolvement','JobSatisfaction','MonthlyIncome','NumCompaniesWorked','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager','business_n','job_n','gender_n','marry_n','overtime']]
```

```python
X.columns
```

```python
y=df.Attrition
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1753)
```

```python
model=LogisticRegression(max_iter=2000,random_state=1)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predict_proba=model.predict_proba(X_test)
```

```python


y_predict=model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.roc_auc_score(y_test,y_predict_proba[:,1])
```

```python
fpr,tpr,thresholds=metrics.roc_curve(y_test,y_predict_proba[:,1])
```

```python
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],color='black',linestyle='--')
plt.grid(True)
plt.show()
```
